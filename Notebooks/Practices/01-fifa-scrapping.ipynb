```python
# tools

import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
product_name_list = []
price_list = []
rating_list = []
sponser_list = []
sold_list = []
product_link =[]

for i in range(1,10):
    url = "https://www.ebay.com/b/Apple/bn_21819543?_pgn="+str(i)
    # url_2 = 'https://www.ebay.com/b/Apple/bn_21819543?_pgn=2'
    print("url :",url)
    page = requests.get(url)
    soup = BeautifulSoup(page.content ,'html.parser')



    product_box = soup.find_all('li',class_= 's-item s-item--large s-item--bgcolored')

    for box in product_box:
    #     print(box.prettify())
    
        link_product = box.find('a')
        
#         print(link_product.get("href"))
        product_name = box.find('h3' ,class_="s-item__title s-item__title--has-tags")
        if product_name:
            pass
        else:
    #         print("*****************")
            product_name= box.find('h3' ,class_="s-item__title")
#             print(product_name.find('span' ,class_="BOLD"))



        sponsership = box.find("div", class_ = "s-item__title-tag")


        price = box.find('span' ,class_ = "s-item__price")
        rate = box.find('span' ,class_= "clipped")
    #     rate = rate.find('span',class_ =  "clipped")
        sold = box.find('span' ,class_="NEGATIVE")

    #     print("-------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

    #     print("-------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

    #     print("-------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
        try:
    #         print(product_name.get_text().strip())
            product_name_list.append(product_name.get_text().strip())
        except:
            product_name_list.append("Nan")

        try:
            sponser_list.append(sponsership.get_text().strip())
    #         print(sponsership.get_text().strip())
        except:
            sponser_list.append("Nan")


        try:
            price_list.append(price.get_text().strip())

        except:
            price_list.append("Nan")


        try:
            rating_list.append(rate.get_text().strip())

        except:
            rating_list.append("Nan")

        try:
            sold_list.append(sold.get_text().strip())

        except:
            sold_list.append("Nan")
        try:
            product_link.append(link_product.get("href"))
            
        except:
            product_link.append("Not available")

```

```python
## all the entity are equal 


print("product_name lenght : ",len(product_name_list))
print("sponser name lenght : ",len(sponser_list))
print("sold list : ",len(sold_list))
print("rating_list lenght : ",len(rating_list))
print("price list lenght : ",len(price_list) )
print("Link product lenght :",len(product_link))
```

```python
## make the dic for dataframe

ebay  = {
    "Product" : product_name_list,
    "Price" : price_list,
    "Sold" : sold_list,
    "Rating" : rating_list,
    "Sponsership" : sponser_list,
    "link" : product_link
    
}
```

```python
df = pd.DataFrame.from_dict(ebay )
```

```python
df.to_csv("ebay_01.csv", index = False, header = True)
```
