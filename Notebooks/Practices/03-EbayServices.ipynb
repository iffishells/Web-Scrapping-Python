```python
import pandas as pd
import requests
import html5lib
from bs4 import BeautifulSoup
```

```python
# step 1 
 # title 
 # link of the product
 # price of the product
def get_String_html(url):
    
    try:
        r = requests.get(url)
        soup = BeautifulSoup(r.content,'html.parser')
    except:
        print("URl is not found")
        print(soup.ok())
        print(soup.status_code)
        
        

    return soup
url = "https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2047675.m570.l1313&_nkw=watch&_sacat=0"
pass_url = get_String_html(url)
```

```python

products_Details = {
    "product_name" : list(),
    "link_product" : list(),
    "price_of_product" : list()
}  
def get_title(soup):
    # it ill return the list of the title of the product
    
    
    p_name  = soup.find_all("h3",class_ = "s-item__title")
    print("lenght of product :  ",len(p_name))
#     print(product_name)


    ## price 
    price = soup.find_all('span',class_= 's-item__price')
    
    ## link of the current product
    product_links = soup.find_all('a' ,class_= 's-item__link' )
    print("length of the product link : ",len(product_links))
    
    
    for name in p_name:
#         print(link.get_text())
        products_Details["product_name"].append(name.get_text()) 
        
    
    
    
    
    print("length of the price : ",len(price))
    i =1 ;
    for p in price:
        print(i  ," ",p.get_text())
        products_Details["price_of_product"].append(p.get_text()) 
        i = i+1
    
    
    
#     print(product_links)
    
    for link in product_links:
        products_Details["link_product"].append(link['href']) 
        


get_title(pass_url)
# get_product_link(get_String_html(url))
```

```python
df = pd.DataFrame.from_dict (products_Details)
```
